```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
full_data = [train,test]

train.head(20)

```

```python
train['contact']=train['contact'].replace('unknown','cellular')
test['contact']=train['contact'].replace('unknown','cellular')

#train['poutcome']=train['poutcome'].replace('unknown',np.nan)
#test['poutcome']=train['poutcome'].replace('unknown',np.nan)

train['month']=train['month'].replace('apr',4)
test['month']=train['month'].replace('apr',4)
train['month']=train['month'].replace('may',5)
test['month']=train['month'].replace('may',5)
train['month']=train['month'].replace('jun',6)
test['month']=train['month'].replace('jun',6)
train['month']=train['month'].replace('jul',7)
test['month']=train['month'].replace('jul',7)
train['month']=train['month'].replace('aug',8)
test['month']=train['month'].replace('aug',8)
train['month']=train['month'].replace('sep',9)
test['month']=train['month'].replace('sep',9)
train['month']=train['month'].replace('oct',10)
test['month']=train['month'].replace('oct',10)
train['month']=train['month'].replace('nov',11)
test['month']=train['month'].replace('nov',11)
train['month']=train['month'].replace('dec',12)
test['month']=train['month'].replace('dec',12)
train['month']=train['month'].replace('jan',1)
test['month']=train['month'].replace('jan',1)
train['month']=train['month'].replace('feb',2)
test['month']=train['month'].replace('feb',2)
train['month']=train['month'].replace('mar',3)
test['month']=train['month'].replace('mar',3)

LE=LabelEncoder()

labels = ['job','marital','default','housing','loan','education','contact','poutcome']
for label in labels:
    train[label]=LE.fit_transform(train[label])
    test[label]=LE.fit_transform(test[label])
train

std_scale = StandardScaler().fit(train[['age', 'balance','duration','pdays']])
train[['age', 'balance','duration','pdays']] = std_scale.transform(train[['age', 'balance','duration','pdays']])
std_scale = StandardScaler().fit(test[['age', 'balance','duration','pdays']])
test[['age', 'balance','duration','pdays']] = std_scale.transform(test[['age', 'balance','duration','pdays']])

train.head(20)
```

```python
X_train = train.drop(['y','id'],axis=1)
Y_train = train['y']

X_test  = test.drop('id',axis=1)
X_train.shape, Y_train.shape, X_test.shape

clf = RandomForestRegressor(n_estimators=1000,criterion="mse",random_state=1,n_jobs=1)
clf.fit(X_train, Y_train)
result = clf.predict(X_test)

print(result)

submission = pd.DataFrame({
        "id": test["id"],
        "y": result
    })

submission.to_csv("submission.csv", index=False)

features = X_train.columns
importances = clf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(6,6))
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.show()

```
